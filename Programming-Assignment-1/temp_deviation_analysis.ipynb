```python
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_axes([0.1, 0.1, 1.5, 1.5])
title = ax1.set_title("Global temperature from 1880 to 2014", fontsize='large')
ax1.plot(temp2_df['Year'], temp2_df['SHem'], 'r', label='Southern Hemisphere')
ax1.plot(temp2_df['Year'], temp2_df['NHem'], 'b', label='Northern Hemisphere')
ax1.plot(temp2_df['Year'], temp2_df['Glob'], 'y--', label='Global')
ax1.set_xticks(np.arange(1880, 2020, step=10))
ax1.legend()

plt.xlabel('Year', fontsize='large')
plt.ylabel('Deviation', fontsize='large')
```
